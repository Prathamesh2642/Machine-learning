```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import math
```

```python
df=pd.read_csv('03_test_scores.csv')
```

```python
df.head()
```

```python
def gradient_descent(x,y):
    m_cur=b_cur=0
    learning_rate=0.0000
    iterations=10000
    n=len(x)
    for i in range(iterations):
        y_pred=m_cur*x+b_cur
        cost=(1/n)*(sum([val**2 for val in (y-y_pred)]))
        md=-(2/n)*(sum(x*y-y_pred))
        bd=-(2/n)*(sum(y-y_pred))        
        m_cur=m_cur-learning_rate*md
        b_cur=b_cur-learning_rate*bd
        print(f"m:{m_cur},b:{b_cur},cost:{cost},iteration:{i}")
  

gradient_descent(df['math'],df['cs'])
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['math']],df['cs'])
```
