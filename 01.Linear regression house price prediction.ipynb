```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
prices={'area':[2600,3000,3200,3600,4000],'prices':[550000,565000,610000,680000,725000]}
```

```python
df=pd.DataFrame(prices)
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('area in sq. feet')
plt.ylabel('prices in $')
plt.scatter(df.area,df.prices,marker="+",color='red')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.prices)
#reg.fit(df[[independent column]],dependent column)
#Y=MX+C          X                     Y
# M is the slope - that gives the best fit line
```

```python
reg.predict([[3300]])
```

```python
reg.coef_ #gives the slope of the regression or best fit line
```

```python
reg.intercept_ #gives the x intercept of the line or the point where y becomes 0
```

```python
# y=m*x+c
y=135.78767123*3300+180616.43835616432
print(y)
```

```python
# y=m*x+c
y=135.78767123*5000+180616.43835616432
print(y)
```
